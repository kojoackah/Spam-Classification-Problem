```python
import pandas as pd
import string
import re
import nltk

pd.set_option('display.max_colwidth', 100)
stopwords = nltk.corpus.stopwords.words('english')
messages = pd.read_csv('C:/Users/USER/Desktop/spam.csv', encoding='latin-1')
messages = messages.drop(labels = ["Unnamed: 2","Unnamed: 3","Unnamed: 4"], axis=1)
messages.columns = ["label","text"]
messages.head()
```

```python
#putting everything together is cleaner and faster, use it next time.


def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [word for word in tokens if word not in stopwords]
    return text
```

```python
# We use the TfidfVectorizer to convert our text messages to a numeric 
# representation in the form of a document term matrix

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(analyzer=clean_text)
t_tfidf = tfidf_vect.fit_transform(messages['text'])
```

```python
# t_tfidf is a document term matrix
X_features = pd.DataFrame(t_tfidf.toarray())
X_features.head()
```

```python
# Import Random Forest for classification from sklearn

from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import precision_score, recall_score
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_features, messages['label'], test_size=0.2)
```

```python
rf = RandomForestClassifier() # uses the defaults

rf_model = rf.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
y_test.head()
```

```python
precision = precision_score(y_test, y_pred, pos_label = 'spam')
recall = recall_score(y_test, y_pred, pos_label = 'spam')

print('Precision: {} / Recall: {}'.format(round(precision,3 ), round(recall, 3)))
```
